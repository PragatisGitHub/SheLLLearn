```python
import keras
from keras.preprocessing import image
from keras.preprocessing.image import img_to_array
from keras.models import load_model
from keras.layers import Dense, Dropout, Flatten

import matplotlib.pyplot as plt #For Plots
```

```python
model = keras.models.load_model("eminst_digits_dense_model - Bkp1.h5")
```

```python
filepath = "uploads/num4.png"
```

```python
# Load the saved image using Keras and resize it to 28x28 pixels
from keras.preprocessing import image
image_size = (28, 28)
im = image.load_img(filepath, target_size=image_size, grayscale=True)
im
```

```python
# Convert the image to a numpy array 
from keras.preprocessing.image import img_to_array
image = img_to_array(im)
image.shape
image
```

```python
# Scale the image pixels by 255 (or use a scaler from sklearn here)
# image /= 255
image = 255 - (image)
```

```python
# Flatten into a 1x28*28 array 
img = image.flatten().reshape(-1, 28*28)
img.shape
```

```python
plt.imshow(img.reshape(28, 28), cmap=plt.cm.Greys)
```

```python
# Invert the pixel values to match the original data
# img = 1 - img
plt.imshow(img.reshape(28, 28), cmap=plt.cm.Greys)
```

```python
image
```

```python
# Make predictions

print(f"One-Hot-Encoded Prediction: {model.predict(img).round()}")
print(f"Predicted class: {model.predict_classes(img)+1}")
```
